```python
# YOUR CODE HERE
# Create a histogram for carbon dioxide (co2) with different bin sizes
bins = [200, 250, 300, 350, 400, 450, 500]  # Experiment with different bin sizes
for bin_size in bins:
    plt.hist(df['co2'], bins=bin_size, edgecolor='black')
    plt.xlabel('CO2 Emissions (g/mile)')
    plt.ylabel('Frequency')
    plt.title(f"CO2 Distribution (Bin Size: {bin_size})")
    plt.show()
```

```python
# YOUR CODE HERE
# Select two other columns based on your interests
columns = ['highway', 'feScore']  # Replace with your chosen columns

# Create histograms for each chosen column
for col in columns:
    plt.hist(df[col])
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.title(f"{col} Distribution")
    plt.show()
```

```python
# Create a correlation matrix
corr_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Fuel Economy Data Correlation Heatmap")
plt.show()
```

**Interpretation of Heatmap:**

The heatmap displays the correlation coefficients between each pair of features in the dataset. Higher values (closer to 1) indicate stronger positive correlations, while lower values (closer to -1) indicate stronger negative correlations. Values closer to 0 indicate weaker or no linear relationship.
