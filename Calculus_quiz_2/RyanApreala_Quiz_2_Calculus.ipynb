# Task 1: Write a dynamic function to find the derivative of any function f(x)

````python
#TO DO: Create a function that does a derivative for any function
import sympy as sp

def dy_dx(f_str):
    """Calculates the derivative of any given function f(x) as a string.

    Args:
        f_str (str): The string representation of the function f(x).

    Returns:
        str: The string representation of the derivative of f(x).
    """

    x = sp.Symbol('x')  # Define the symbolic variable x
    f = sp.sympify(f_str)  # Parse the input string as a SymPy expression
    dy_dx = sp.diff(f, x)  # Calculate the derivative
    return str(dy_dx)  # Return the derivative as a string
````
